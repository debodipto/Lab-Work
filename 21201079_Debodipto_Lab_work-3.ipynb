```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return 5 * np.sin(2*x)
```

```python
def derivative(f, x, h=0.001):
    return (f(x+h) - f(x)) / (h)
```

```python
true_value = 9.513986399606658
h = [0.3, 0.15, 0.1, 0.05, 0.01, 0.001]
x = 3
for i in h:
    approx_value = derivative(f, x, i)
    print("h = ", i, "; derivative = ", approx_value, "; RTE = ", (abs(true_value - approx_value)) * 100 /true_value, "%")
```

```python
# let the first approximation be for h = 0.3
prev_approx = derivative(f, x=3, h=0.3)

# let the successive approximations be for decreasing values of h
h = [0.15, 0.1, 0.05, 0.01, 0.001, 0.0001]
errors = []
for i in h:
    present_approx = derivative(f, x=3, h=i)
    RAE = abs(present_approx - prev_approx)*100/present_approx
    errors.append(RAE)
    prev_approx = present_approx

print("h values: ", h)
print("errors: ", errors)

# plot the errors
plt.figure(figsize=(10, 5))
plt.plot(h, errors, marker = 'x', color = 'red', linewidth = 3, markersize = 10)
plt.xlabel("h", fontdict={'fontsize': 15})
plt.ylabel("RAE", fontdict={'fontsize': 15})
plt.title("Relative Approximate Error vs Step Size", fontdict={'fontsize': 20})
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function f(x) = 5*sin(2x)
def f(x):
    return 5 * np.sin(2 * x)

# Define the true derivative of f(x) = 5*sin(2x)
def f_prime(x):
    return 10 * np.cos(2 * x)  # Derivative of 5*sin(2x) is 10*cos(2x)

# True derivative at x = 3
true_value = f_prime(3)

# List of h values
h_values = [0.3, 0.15, 0.1, 0.05, 0.01, 0.001, 0.0001]
true_errors = []

for h in h_values:
    # Numerical derivative using central difference formula
    present_approx = derivative(f, x, h)

    # True Error: absolute difference between true value and numerical approximation
    true_error = abs(true_value - present_approx)
    true_errors.append(true_error)

# Print h values and corresponding true errors
print("h values: ", h_values)
print("True Errors: ", true_errors)

# Plot the true errors vs. h
plt.figure(figsize=(10, 5))
plt.plot(h_values, true_errors, marker='o', color='purple', linewidth=3, markersize=10)
plt.xlabel("h", fontdict={'fontsize': 15})
plt.ylabel("True Error", fontdict={'fontsize': 15})
plt.title("True Error vs Step Size (h)", fontdict={'fontsize': 20})
# plt.xscale('log')  # Using logarithmic scale for better visualization
# plt.yscale('log')
plt.grid(True)
plt.show()
```
